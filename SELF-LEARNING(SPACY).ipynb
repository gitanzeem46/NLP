```python
# Adding special case Tokenization rule
```

```python
from spacy.symbols import ORTH
```

```python
# Load the English model
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp ("tanzeem that")
doc
```

```python
for token in doc:
  print(token.text)
```

```python
special_case = [{ORTH:"tan"},{ORTH:"zeem"}]
special_case
```

```python
nlp.tokenizer.add_special_case("tanzeem",special_case)
```

```python
for token in nlp ("tanzeem that"):
  print(token.text)
```

```python
# 2nd Methods
nlp.vocab["what"].is_stop
```

```python
doc = nlp ("pyhton is a programming language. i am learning natural language processing")
doc
```

```python
for token in doc:
  print(token.text)
```

```python
for token in doc:
  if nlp.vocab[token.text].is_stop:
    print(token.text)
```

```python
for token in doc:
  if not nlp.vocab[token.text].is_stop:
    print(token.text)
```
